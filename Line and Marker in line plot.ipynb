```python
import matplotlib.pyplot as plt # pyplot= pie plot
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
a=[2,5,8]
b=[4,6,10]
plt.plot(a,b,color="green",marker="^",linestyle="--") # We can also use inside of linestyle = -,--,-.,:
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line chart")
plt.show()
```

```python
a=[2,5,8]
b=[4,6,10]
plt.plot(a,b,color="green",marker="o",linestyle="--",markersize=14) # We can also use inside of linestyle = -,--,-.,:
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line chart")
plt.show()
```

```python
a=[2,5,8]
b=[4,6,10]
plt.plot(a,b,color="green",marker="^",linestyle="--"); # We can also use inside of linestyle = -,--,-.,:
plt.xlabel("x axis");
plt.ylabel("y axis");
plt.title("line chart");
# plt.show()
```

# We Can also use color , marker , linestyle this type

```python
a=[2,5,8]
b=[4,6,10]
plt.plot(a,b,'go--'); # We can also use for creating the color,marker,linespace (g=green , o = marker , -- = linestyle)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line chart")
plt.show()
```

```python
a=[2,5,8]
b=[4,6,10]
plt.plot(a,b,'bo:'); # We can also use for creating the color,marker,linestyle (b = blue , o = marker , : = linestyle)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line chart")
plt.show()
```

```python
a=np.arange(1,10)
b=np.arange(10,19)
plt.plot(a,b,color="green",marker="^",linestyle="--");
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line chart")
plt.show()
```

```python
l1=np.linspace(1,10,5)
l2=np.linspace(10,20,5)
plt.plot(l1,l2,color="red",marker="o",linestyle="--",markersize=14);
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line chart")
plt.show()
```

```python
l2=l1**3
plt.plot(l1,l2,color="green",marker="o",linestyle="--",markersize=14);
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
a=[1,2,4,5,6]
b=[7,8,9,10,11]
plt.figure(figsize=(10,4),dpi=100)
plt.plot(a,b,color="red",marker="o",linestyle=":",markersize=14,label="Salary")
plt.legend()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Chart")
plt.show()
```
