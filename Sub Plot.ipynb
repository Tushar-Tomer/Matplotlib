```python
import matplotlib.pyplot as plt # pyplot= pie plot
import seaborn as sns
import numpy as np
import pandas as pd
```

# Sub plot
## Subplot is used to show the multiple plots in one figure.
## We ca use this method to separate two graphs which plotted in the same axis Matplotlib supports all kinds of subplots , including 2x1 vertical , 2x1 horizontal , or a 2x2 grid.
## It accepts the three arguments.
## They are nrows , ncols , and index.
## It denotes the number of rows , number of columns , and the index.

```python
year=["2013","2014","2015","2016","2017","2018","2019","2020","2021","2022"]
online_sale=[5,8,8,9,10,12,8,7,20,22]
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.bar(year,online_sale)
plt.xlabel("year")
plt.ylabel("online_sale")
plt.title("maximum sale analyses")
plt.xticks(rotation=90)
plt.subplot(1,2,1)
plt.plot(year,online_sale)
plt.xlabel("year")
plt.ylabel("online_sale")
plt.title("sale trend analyses")
plt.xticks(rotation=90)
plt.show()

```

```python
year=["2013","2014","2015","2016","2017","2018","2019","2020","2021","2022"]
online_sale=[5,8,8,9,10,12,8,7,20,22]
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(year,online_sale)
plt.xlabel("year")
plt.ylabel("online_sale")
plt.title("maximum sale analyses")
plt.xticks(rotation=90)
plt.subplot(1,2,2)
plt.plot(year,online_sale)
plt.xlabel("year")
plt.ylabel("online_sale")
plt.title("sale trend analyses")
plt.xticks(rotation=90)
plt.show()

```
