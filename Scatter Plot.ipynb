```python
x = np.array([1,2,5,5,10,11,9,10,4,7,6,8,6])  ## expererince in years
y = np.array([4,5,9,9,10,10,9,10,7,6,8,8,8])  ### salary packages in lacs

plt.scatter(x, y,color="g")
plt.xlabel("Years")
plt.ylabel("Salary in lakhs")
plt.show()


```

```python
x = np.array([2,4,8,10,11,10,11,13,14,15,16,18,20])  ## expererince in years

y = np.array([4,8,8.5,10,12,12,13,13,14,15,16,18,12])  ### salary packages in lacs
z=  np.array([4.1,5.2,9.1,9.0,10.0,10.1,10.2,10.2,10.2,10.0,10.0,10.0,11])  ## cgpa
plt.figure(figsize=(10,4))
plt.scatter(y,x,color="r",label="Experience")
plt.scatter(y,z,label="Marks")
plt.xlabel("Salary")
plt.ylabel("Experience & Marks")
plt.legend()
plt.show()

```
