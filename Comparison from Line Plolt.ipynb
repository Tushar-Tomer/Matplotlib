```python
import matplotlib.pyplot as plt # pyplot= pie plot
import seaborn as sns
import numpy as np
import pandas as pd
```

# Comparisons

```python
a=np.linspace(1,10,5)
b=np.linspace(10,20,5)
plt.figure(figsize=(5,3),dpi=140) # Figure is large with the help of dpi
x=b**2
y=b**3
plt.plot(a,b,label="2010",marker=">")
plt.plot(a,x,label="2020",marker=">")
plt.plot(a,y,label="2022",marker=">")
plt.title("Sales")
plt.legend()
plt.show()
```

```python
a=np.linspace(1,10,5)
b=np.linspace(10,20,5)
plt.figure(figsize=(5,3),dpi=140) # Figure is large with the help of dpi
x=b**2
y=b**3
plt.plot(a,b,label="2010",marker=">",color="r",markersize=10)
plt.plot(a,x,label="2020",marker=">",color="y",markersize=10)
plt.plot(a,y,label="2022",marker=">",color="g",markersize=10)
plt.title("Sales")
plt.legend()
plt.show()
```

# Find out the year in which number of passengers traveled most and also tell me whether the no. of passenger over the time increasing and decreasing

```python
a["year"].unique()
```

```python
x=a.groupby("year")["passengers"].sum().reset_index()
x
```

```python
plt.plot(x["year"],x["passengers"],color="r",marker="o",markersize=8)
plt.xticks(rotation=45)
plt.xlabel("Years")
plt.ylabel("Passengers")
plt.title("Passenger Information")

plt.show()
```

```python
plt.figure(figsize=(10,4),dpi=80)
plt.bar(x["year"],x["passengers"],color="g")
plt.xticks(rotation=45)
plt.xlabel("Years")
plt.ylabel("Passengers")
plt.show()
```

```python
# Highest passengers travelled in 1960 and passengers are increase over the time.
```
