```python
import matplotlib.pyplot as plt # pyplot= pie plot
import seaborn as sns
import numpy as np
import pandas as pd
```

# Pie Chart 
# Pie chart is used for univarient analyses it is used for categorical type of data when we want to see the distribution of the categories in the form of percent. 

```python
# Use nothing
```

```python
l=[44,22,88,54]
plt.pie(l);
```

```python
# Use labels
```

```python
l=[44,22,88,54]
plt.pie(l,labels=["C++","C","Python","Java"]);
```

```python
# autopct
# Use labels , autopct(percentage)
```

```python
l=[44,22,88,54]
plt.pie(l,labels=["C++","C","Python","Java"],autopct="%0.02F%%"); # autopct always gives in string
```

```python
l=[44,22,88,54]
plt.pie(l,labels=["C++","C","Python","Java"],autopct="%0.01f%%"); # 0.0 ka baad hum jitney no. deyenge utni points ma percent milegi
```

```python
# colors
# Use labels , autopct(percentage) , colors
```

```python
l=[44,22,88,54]
plt.pie(l,labels=["C++","C","Python","Java"],autopct="%0.01f%%",colors=["orange","red","green","yellow"]);
```

```python
# shadow
# Use labels , autopct(percentage) , colors , shadow
```

```python
l=[44,22,88,54]
plt.pie(l,labels=["C++","C","Python","Java"],autopct="%0.01f%%",colors=["orange","red","green","yellow"],shadow=True);
```

```python
# explode
# Use labels , autopct(percentage) , colors , shadow , explode
```

```python
l=[44,22,88,54]
plt.pie(l,labels=["C++","C","Python","Java"],autopct="%0.01f%%",colors=["orange","red","green","yellow"],shadow=True,explode=[0,0,1,0]); # Explode ma particular data ka liye jitney no. hum deyenge uthi dur vo data ho jayega
```

```python
# Use All
```

```python
l=[200,300,310,800]
plt.pie(l,labels=["python","ds","ba","st"],autopct="%0.00f%%",explode=[0,0,5,0],colors=["red","green","pink","blue"],shadow=True);
```
