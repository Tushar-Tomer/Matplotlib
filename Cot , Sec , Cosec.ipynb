```python
import numpy as np
import matplotlib.pyplot as plt
```

# cot()  (cot use karne ka liye phale tan nikalna hai aur uske baad arctan ) (cot() = 1/tan)

```python
import numpy as np
a=np.arange(1,10)
b=np.tan(a)
cot=np.arctan(b)
cot
```

```python
plt.plot(cot);
```

# sec()  (sec use karne ka liye phale cos nikalna hai aur uske baad arccos ) (sec() = 1/cos)

```python
a=np.arange(1,10)
b=np.cos(a)
sec=np.arccos(b)
sec
```

```python
plt.plot(sec);
```

# cosec()  (cosec use karne ka liye phale sin nikalna hai aur uske baad arcsin ) (cosec() = 1/sin)

```python
a=np.arange(1,10)
b=np.sin(a)
cosec=np.arcsin(b)
cosec
```

```python
plt.plot(cosec);
```
