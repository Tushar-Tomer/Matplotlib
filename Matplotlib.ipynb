# We can use this function/keyword in matplotlib
## 1. color = "Enter Color Name"
## 2. marker = "Enter marker sign"  (^ , o , etc)
## 3. linestyle = "Enter linestyle" (- , -- , -. , :)
## 4. markersize = Enter No.
## 5. plt.plot(variable,variable)
## 6. plt.xlabel("x axis")
## 7. plt.ylabel("y axis")
## 8. plt.title("Enter Title name")
## 9. plt.show() / ;(semicolen)
## 10. label="Enter lable name" ( plt.legend() lagana compulsary hai label dene ka liye )
## 11. figure size = plt.figure(figsize=(Enter lenght , Enter Height))    
## Note : figure size plot karne se phale deyenge.
## 12. dpi : dpi=Enter figure size no.  (dpi large the figure size)
## 13. rotate words = plt.xticks(rotation=Enter No.)  ( This keyword is used for rotating x axis )
## 14. change ticks(x labels) = plt.xticks([Enter label numbers]) ( With the help of this keyword we will gives the x axis labels )
## 15. width : width=.2/.1/.4/.5 (width change the size of bar line by default it .5)
## 16. edgecolor="Enter color name" (edgecolor change the color of bar edge).
## 17. piechart labels = labels=["Enter Label Name"]
## 18. autopct="%0.01f%%" (It gives the percentage and 0.0 ka baad jitnay no. hum deyenge utna no. tak hama points ma answer milega).
## 19. shadow=True (It gives shadow in pie chart)
## 20. explode=[0,0,0,0] ( jis bhi data ko alag karna hai use data pr no. deyenge aur no. hama jitni door karna hai utna no. deyenge)

# Sub plot
## Subplot is used to show the multiple plots in one figure.
## We ca use this method to separate two graphs which plotted in the same axis Matplotlib supports all kinds of subplots , including 2x1 vertical , 2x1 horizontal , or a 2x2 grid.
## It accepts the three arguments.
## They are nrows , ncols , and index.
## It denotes the number of rows , number of columns , and the index.

```python
year=["2013","2014","2015","2016","2017","2018","2019","2020","2021","2022"]
online_sale=[5,8,8,9,10,12,8,7,20,22]
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.bar(year,online_sale)
plt.xlabel("year")
plt.ylabel("online_sale")
plt.title("maximum sale analyses")
plt.xticks(rotation=90)
plt.subplot(1,2,1)
plt.plot(year,online_sale)
plt.xlabel("year")
plt.ylabel("online_sale")
plt.title("sale trend analyses")
plt.xticks(rotation=90)
plt.show()

```

```python
year=["2013","2014","2015","2016","2017","2018","2019","2020","2021","2022"]
online_sale=[5,8,8,9,10,12,8,7,20,22]
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(year,online_sale)
plt.xlabel("year")
plt.ylabel("online_sale")
plt.title("maximum sale analyses")
plt.xticks(rotation=90)
plt.subplot(1,2,2)
plt.plot(year,online_sale)
plt.xlabel("year")
plt.ylabel("online_sale")
plt.title("sale trend analyses")
plt.xticks(rotation=90)
plt.show()

```

# Find percentile , quantile , lower limit , upper limit

# Percentile tells the proportion of population that falls below a value .

```python
# finding upper limit and lower limit
```

```python
np.percentile(a["tip"],[25,50,75])
```

```python
q1=a["tip"].quantile(0.25)
q1
```

```python
q3=a["tip"].quantile(0.75)
q3
```

```python
iqr=q3-q1 # IQR = Inter quantile range
iqr
```

```python
upper_limit=q3+1.5*iqr
upper_limit
```

```python
lower_limit=q1-1.5*iqr
lower_limit
```

```python
a[a["tip"]<lower_limit]
```

```python
a[a["tip"]>upper_limit]
```

```python
b=a[(a["tip"]>lower_limit)&(a["tip"]<upper_limit)]# 
b
```

```python
d=a.copy()
d
```

```python
d["tip"]=np.where(d["tip"]>upper_limit,upper_limit,np.where(d["tip"]<lower_limit,lower_limit,a["tip"]))
d["tip"]
```
